```python
from owslib.wms import WebMapService
import shapefile
import json
import os.path
from os import listdir
```

```python
#Listamos los ficheros .shp de la carpeta ./shape
archivos=listdir("./shape")
for archivo in archivos:
    nombre, extension = os.path.splitext(archivo)
    if (extension=='.shp'):
        
        #leemos fichero shape
        sf = shapefile.Reader("./shape/"+archivo)

        #URL WMS de PNOA
        URL = "https://www.ign.es/wms-inspire/PNOA-MA?SERVICE=WMS&REQUEST=GETCAPABILITIES"
        wms = WebMapService(URL, version='1.1.1')

        #Fichero salida imagenes PNOA 
        OUTPUT_DIRECTORY = './images/imagenes/'

        #Coordenadas UTM donde empieza la descarga
        x_min=round(sf.bbox[0],2)
        y_min=round(sf.bbox[1],2)

        #dx y df es la distancia en metros entre fotografias
        #Debe ser el tama√±o fotografia (size) por 0.25 (resolucion PNOA 0.25 cms/pixel)
        dx, dy = 96, 96

        #Mosaico de imagenes
        no_tiles_x = 10
        no_tiles_y = 10

        #Total imagenes mosaico
        total_no_tiles = no_tiles_x * no_tiles_y

        #Coordenadas maximas XY
        x_max = x_min + no_tiles_x * dx
        y_max = y_min + no_tiles_y * dy

        #Cuadro contenedor imagenes
        BOUNDING_BOX = [x_min, y_min, x_max, y_max]

        #Extraccion WMS
        #Recorremos las x
        for ii in range(0,no_tiles_x):
            #recorremos las y
            for jj in range(0,no_tiles_y):
                #Coordenada x de la imagen
                ll_x_ = x_min + ii*dx
                #Coordenada y de la imagen
                ll_y_ = y_min + jj*dy
                #Cuadro delimitador de la imagen
                bbox = (ll_x_, ll_y_, ll_x_ + dx, ll_y_ + dy) 
                #obtenemos la imagen
                img = wms.getmap(layers=['OI.OrthoimageCoverage'], srs='EPSG:25830', 
                                 bbox=bbox, size=(384, 384), format='image/jpeg', transparent=True)
                #Nombre del fichero contiene las coordendas de la imagen
                filename = "{}_{}_{}_{}.jpg".format(bbox[0], bbox[1], bbox[2], bbox[3])
                #Guardamos la cimagen
                out = open(OUTPUT_DIRECTORY + filename, 'wb')
                out.write(img.read())
                out.close()
        print("Capa finalizada " + archivo)
print("Proceso finalizado ")
```
